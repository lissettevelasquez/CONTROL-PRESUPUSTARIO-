```python
# Cargar el archivo
file_path =  '/content/drive/MyDrive/CP-COQ-09.xlsx'
real_acum_df = pd.read_excel(file_path, sheet_name="REAL")
ppto_acum_df = pd.read_excel(file_path, sheet_name="PPTO")
```

```python
# Renombrar las columnas para estandarizar
real_acum_df.rename(columns={
    "Año": "año",
    "Mes": "mes",
    "Area Negocio Desc": "area negocio",
    "Monto Real": "monto real",
    "Grupo Cuenta Desc": "grupo cuenta desc",
    "Centro Costo Nvl2 Desc": "centro costo nvl2 desc",
    "Cuenta Cod Desc Larga": "cuenta cod desc larga",
    "Centro Costo Desc": "centro costo desc"
}, inplace=True)

ppto_acum_df.rename(columns={
    "Año": "año",
    "Mes": "mes",
    "Area Negocio Desc": "area negocio",
    "Monto Presupuestado": "Monto presupuestado",
    "Grupo Cuenta Desc": "grupo cuenta desc",
    "Centro Costo Nvl2 Desc": "centro costo nvl2 desc",
    "Cuenta Cod Desc Larga": "cuenta cod desc larga",
    "Centro Costo Desc": "centro costo desc"
}, inplace=True)
```

```python
# Seleccionar columnas relevantes
real_acum_selected = real_acum_df[["año", "mes", "area negocio", "monto real", "grupo cuenta desc",
                                   "centro costo nvl2 desc", "cuenta cod desc larga", "centro costo desc"]]

ppto_acum_selected = ppto_acum_df[["año", "mes", "area negocio", "Monto presupuestado", "grupo cuenta desc",
                                   "centro costo nvl2 desc", "cuenta cod desc larga", "centro costo desc"]]
```

```python
# Combinar las tablas
combined_df = pd.merge(
    real_acum_selected,
    ppto_acum_selected,
    on=["año","mes", "area negocio", "grupo cuenta desc", "centro costo nvl2 desc", "cuenta cod desc larga", "centro costo desc"],
    how="outer"
)
```

```python
# prompt: mostrar resultado en formato de tabla (solo las primeras cinco filas)

print(combined_df.head())
```

```python
# prompt: guardar tabla combinada en un archivo excel

# Guardar la tabla combinada en un nuevo archivo Excel
output_file_path = '/content/drive/MyDrive/CP-sep-COQ.xlsx'
combined_df.to_excel(output_file_path, index=False)  # index=False evita guardar el índice del DataFrame

print(f"Tabla combinada guardada en: {output_file_path}")
```
